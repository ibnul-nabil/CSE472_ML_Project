
This notebook demonstrates how to use the trained Bengali TTS model to synthesize speech from text and evaluate its output quality.

1. Import Required Libraries
Import necessary libraries for text-to-speech (TTS), audio processing, and playback.

```python
# Install required packages if not already installed
# !pip install TTS==0.22.0 IPython
# !sudo apt-get install espeak-ng -y

import torch
from TTS.utils.manage import ModelManager
from TTS.utils.synthesizer import Synthesizer
from IPython.display import Audio, display
import numpy as np
import scipy.io.wavfile as wav
import os
import json
```

2. Load the best_model TTS Model
Load the pre-trained Bengali TTS model and its configuration.

```python
# Paths to model and config
MODEL_PATH = "./best_model_141322_yt2.pth"
CONFIG_PATH = "./config.json"

# Check if files exist
assert os.path.isfile(MODEL_PATH), f"Model file not found: {MODEL_PATH}"
assert os.path.isfile(CONFIG_PATH), f"Config file not found: {CONFIG_PATH}"

# Load config
with open(CONFIG_PATH, "r") as f:
    config = json.load(f)

# Initialize synthesizer
synthesizer = Synthesizer(
    tts_checkpoint=MODEL_PATH,
    tts_config_path=CONFIG_PATH,
    use_cuda=torch.cuda.is_available()
)
print("Model loaded successfully.")
```

## 3. Prepare Test Text Inputs
Define a set of Bengali text samples to test the TTS model's capabilities.

```python
# Example Bengali text samples for TTS testing
bengali_texts = [
    "তাই আশেপাশে কোথাও মাছের কথা শুনলেই সেখানে আমরা ছুটে যাই মাছ ধরার জন্য",
    "গত কয়েক দিন আগে আমাদের পাশের এলাকা থেকে আমাদেরকে দাওয়াত দেওয়া হয়েছিল মাছ ধরার জন্য অর্থাৎ বিলে পলো দিয়ে মাছ ধরার জন্য",
    "আসসালামু আলাইকুম আসকে আমরা একটা বড় সিদ্ধান্ত নিয়েছি সেইটা হুচ্ছে আমাদের যে গুর হয় সেই গুরের সমস্ত রস আসকে আমরা নিয়ে আসব",
    "ডিএনএ মেরামত ও ক্যান্সারের জেনেটিক্স সম্পর্কে অনেকগুলি গবেষণাপত্রে তাঁর অবদান রয়েছে",
    "তাই এর পর থেকে এধরনের তারার সমষ্টিকে সাধারণভাবে গ্যালাক্সি বা ছায়াপথ নামে অভিহিত করা হতে থাকে"
]

# Optionally, add more diverse or challenging sentences as needed.
```

 4. Synthesize Speech from Text
Use the model to convert each test text input into audio waveforms.

```python
# Synthesize speech for each text input
outputs = []
for idx, text in enumerate(bengali_texts):
    print(f"Synthesizing {idx+1}/{len(bengali_texts)}: {text}")
    wavs = synthesizer.tts(text)
    outputs.append(wavs)
print("Synthesis complete.")
```

 5. Play or Save Synthesized Audio
Play the generated audio directly in the notebook for each test input.

```python
# Play the generated audio for each text
for idx, (text, wavs) in enumerate(zip(bengali_texts, outputs)):
    print(f"Text {idx+1}: {text}")
    display(Audio(wavs, rate=synthesizer.output_sample_rate))
```

```python
for idx, (text, wav_data) in enumerate(zip(bengali_texts, outputs)):
    output_path = f"./inference/output{idx+1}.wav"
    wav_array = np.array(wav_data)
    wav.write(output_path, synthesizer.output_sample_rate, wav_array)
    print(f"Saved {output_path} | Text: {text}")
```
